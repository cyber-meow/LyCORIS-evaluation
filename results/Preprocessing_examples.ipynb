```python
def compute_scaled_ranks(df,
                         groupby_columns,
                         score_metrics=[],
                         distance_metrics=[]):
    """
    Compute scaled ranks within groups defined by groupby_columns.

    Parameters:
    - df: DataFrame containing the data.
    - groupby_columns: List of column names used to group the data.
    - score_metrics: List of metrics where a higher score is better.
    - distance_metrics: List of metrics where a lower score is better.

    Returns:
    - DataFrame with scores in specified columns replaced by scaled ranks.
    """
    def rank_score_metric(group, metric):
        """Computes the scaled rank for score metrics."""
        return group[metric].rank(ascending=False) / len(group)

    def rank_distance_metric(group, metric):
        """Computes the scaled rank for distance metrics."""
        return group[metric].rank(ascending=True) / len(group)

    for col in df.columns:
        if col[0] in score_metrics:
            df[col] = df.groupby(groupby_columns).apply(
                lambda g: rank_score_metric(g, col))

        elif col in distance_metrics:
            df[col] = df.groupby(groupby_columns).apply(
                lambda g: rank_distance_metric(g, col))

    return df
```

```python
def compute_scaled_ranks(df,
                         groupby_columns,
                         score_metrics=[],
                         distance_metrics=[]):
    """
    Compute scaled ranks within groups defined by groupby_columns.

    Parameters:
    - df: DataFrame containing the data.
    - groupby_columns: List of column names used to group the data.
    - score_metrics: List of metrics where a higher score is better.
    - distance_metrics: List of metrics where a lower score is better.

    Returns:
    - DataFrame with scores in specified columns replaced by scaled ranks.
    """
    
    def rank_score_metric(group):
        """Computes the scaled rank for score metrics."""
        return group.rank(ascending=False) / len(group)

    def rank_distance_metric(group):
        """Computes the scaled rank for distance metrics."""
        return group.rank(ascending=True) / len(group)

    for col in df.columns:
        if col.split('|')[0] in score_metrics:
            df[col] = df.groupby(groupby_columns)[col].transform(rank_score_metric)
        elif col.split('|')[0] in distance_metrics:
            df[col] = df.groupby(groupby_columns)[col].transform(rank_distance_metric)

    return df
```
