### Reweight

```python
metrics = pd.read_csv('ddim_sd15_metrics.csv')
metrics_reweight = pd.read_csv('reweight_metrics.csv')
```

```python
selected = ['1001', '1002', '1011', '1012', '1021', '1022', '1031']
condition = metrics['Folder'].apply(lambda x: any(keyword in x for keyword in selected))
metrics_filtered = metrics[condition]
```

```python
metrics_reweight_update = pd.concat([metrics_filtered, metrics_reweight])
display(metrics_reweight_update)
```

```python
with display_all_rows():
    display(metrics_reweight_update.isna().any())
```

```python
metrics_reweight_update.to_csv('reweight_metrics.csv', index=False)
```
